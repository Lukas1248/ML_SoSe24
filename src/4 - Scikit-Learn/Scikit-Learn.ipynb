## 1. Data Preprocessing in Machine Learning

Data preprocessing is a process of preparing the raw data and making it suitable for a machine learning model. It is the first and crucial step while creating a machine learning model.

When creating a machine learning project, it is not always the case that we come across a clean and formatted data. And while doing any operation with data, it is mandatory to clean it and put in a formatted way. So for this purpose, we use the data preprocessing task.

### Why do we need Data Preprocessing?
Real-world data generally contains noises and missing values. It can also be in an unusable format which cannot be directly used for machine learning models. Data preprocessing is required for `cleaning the data` and `making it suitable for a machine learning model` which also increases the accuracy and efficiency of the deployed model.

#### Example DataFrame for data preprocessing
On the next sections, we will use the following DataFrame to discuss the contents of data preprocessing:

```python
#ColumnNames = ['Country','Age','Salary','Purchased']
Countries = ['Belgium','France','Germany','France','Germany','Belgium','Germany','France','Belgium','France']
Ages = [38, 43, 30, 48, float('NaN'), 35, float('NaN'), 49, 50, 37]
Salaries = [68000, 45000, 54000, 65000, 56000, 58000, float('NaN'), 79000, 88000, 77000]
Purchased = ['no','yes','no','yes','no','no','yes','no','yes','yes']
aDict = {'Country' : Countries, 'Age' : Ages, 'Salary' : Salaries, 'Purchased' : Purchased}

# create dataframe
df = pd.DataFrame(data=aDict)

# make a copy of the dataframe
data = df.copy()

# display dataframe
data
```

### 1.1. Handling missing values

If the dataset contains some missing data, then it may create a huge problem for the machine learning model. Hence it is necessary to handle missing values present in the dataset. Missing values can be identified as `None` (null) or `NaN` (Not a Number). To check a DataFrame for missing values, we can use the `isnull()` method of pandas.

For example:

```python
# calculate the sum of missing values for each attribute
data.isnull().sum()
```

After we spot missing values in our DataFrame, we have to choose between two ways to handle missing data:
- deleting the particular rows of the dataset which contain missing data
- imputing (replacing) the missing data

#### Deleting particular rows
In this way, we just delete the specific row or column which consists of missing values. But this way is not so efficient and removing data may lead to loss of information which will not give the accurate output.

#### Imputing missing data
In this way, we will impute the missing data with interpolation or statistical features (e.g. mean, median ...) of the existing data.
This strategy is useful for the attributes which have numeric data such as age, salary, year, etc. 

To handle missing values, we will use the `imputer` method of the Scikit-Learn package. Below you can find an example for imputing missing values in the `Age` and the `Salary` column of our dataset.

```python
# handling missing data (replacing missing data with the median)
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='median')  

# fitting imputer object to the independent variables x.
# When you run imp.fit, it calculates the value to replace in each column (in this case the median)
imp.fit(data[['Age','Salary']])  

# to access the values which will be used to replace the missing values in each column:
print(imp.statistics_)

# replacing missing data with the median of the attribute
data[['Age','Salary']] = imp.transform(data[['Age', 'Salary']])

```

```python
# After imputation of the missing values
data[['Age', 'Salary']]
```

### 1.2. Encoding categorical data

Categorical data is data which has some categories. In our dataset example, there are two categorical attributes: `Country` and `Purchased`.

Since machine learning models completely work on mathematics and numbers, we need to convert our categorical information in numerical information. In this case it is necessary to encode categorial variables into numbers. We can use the following encoding types to convert categorical information into numerical information:
- integer encoding
- one-hot encoding

#### Integer encoding
Integer encoding is a method where each categorical value is assigned an integer value. This is called a label encoding or an integer encoding and is easily reversible. The integer values have a **natural ordered relationship between each other and machine learning algorithms may be able to understand and harness this relationship.**

We can use the `LabelEncoder()` method of the Scikit-Learn library for integer encoding.

In the case of the `Purchased` attribute of our dataset, we have two options (*yes*, *no*) which represent categorical information. We can encode yes and no to integer values by using the `LabelEncoder()` method as following:

```python
# Categorical data for Country attribute  
from sklearn.preprocessing import LabelEncoder  

label_encoder_x = LabelEncoder()  
#Fit label encoder and return encoded labels
data['Purchased'] = label_encoder_x.fit_transform(data['Purchased']) 

# Alternative: you can perform the fitting and transform in two separate steps
# like following:
#label_encoder_x.fit(data['Purchased'])  
#data['Purchased'] = label_encoder_x.transform(data['Purchased'])  
```

```python
# before encoding of the categorical values
df['Purchased']
```

```python
# after encoding of the categorical values
data['Purchased']
```

#### One-hot encoding

For categorical variables where no ordinal relationship exists, the integer encoding is not enough. In fact, using integer encoding and allowing the model to assume a natural ordering between categories may lead to poor performance or unexpected results.

In this case, a one-hot encoding can be applied to the integer representation. This is where the integer encoded variable is removed and a new binary variable is added for each unique integer value.

For example, we can one-hot encode the `Country` attribute of our dataframe as following:

```python
# one-hot encoding of the Country column
x = pd.get_dummies(data['Country'], prefix='', prefix_sep='')
x
```

```python
# concaternate new one hot encoded DataFrame into our main DataFrame
data = pd.concat([data, x], axis=1)
# display DataFrame
data
```

```python
# delete the Country column of our DataFrame after one-hot encoding
data = data.drop(columns=['Country'])
# display DataFrame
data
```

### 1.3. Splitting dataset into training and test set

In machine learning data preprocessing, we divide our dataset into a `training set` and `test set`. This is one of the crucial steps of data preprocessing as by doing this, we can evaluate the performance of our machine learning model.

Only data that the model has not seen is relevant for the evaluation of a machine learning model. It is therefore important to subdivide the existing dataset into a training and a test set before training the model.

In this context, the `training set` describes a subset of the dataset which is used to train the machine learning model. The `test set` on the other hand is a subset of the dataset which is only used for the evaluation of the trained model. The test set contains new and unseen information for the machine learning model.

The division of the data into the training and the test set can differ depending on the amount of data available and the machine learning model used. The following cases can be used as a guide:
- low amount of data (~ 10 thousands) $\to$ train set: 70 %, test set = 30 %
- high amount of data (> 100 thousands) $\to$ train set: 99 %, test set = 1 %

There are many ways to split the data into training and test sets. Scikit-Learn provides two main functions for this purpose, which are embedded in the `model_selection` module of the library:
- `train_test_split()` and

- `StratifiedShuffleSplit()`

The `train_test_split()` method splits the dataset into a random training and a test set determined by the argument `test_size`. The argument `shuffle` determines if the data should be randomly shuffled. 

The `StratifiedShuffleSplit()` method also splits the dataset into a random training and test set but keeps the ratio between the individual attributes of the train and test sets the same as in the original dataset. `StratifiedShuffleSplit()` is mostly used when the dataset contains `categorical attributes` to make sure that the separation of the categorial attributes values to the training and test set is respresentative of the overall population.

You can use the `train_test_split()` method as following:

```python
# Simple example:
from sklearn.model_selection import train_test_split
X, y = np.arange(10).reshape((5, 2)), range(5)
print('X \n', X)
list(y)
print('\n')
print('y \n', y)
print('\n')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42,shuffle=True)
print('X_train \n',X_train)
print('\n')
print('y_train \n', y_train)
# Change Shuffle to False one time and delete random_state=42 to see the difference 
```

### 1.4. Feature scaling
Feature scaling is a technique to standardize the independent variables of the dataset in a specific range. In feature scaling, we put our variables in the same range and in the same scale so that no any variable dominates the other variables.

Example:


As we can see, the `Age` and the `Salary` column values are not on the same scale. Depending on the selection of a machine learning model, the large scale gap between these attributes could cause big issues. For example, if we would used a machine learning model which is based on the euclidean distance between data points in a two dimensional feature space, the attribute `Salary` would dominate the attribute `Age` leading to flawed results.

The euclidian distance *`euc`* between two data points 1 and 2 with the attributes x and y is calculated by:

$ euc = \sqrt{(x_2 - x_1)^2+(y_2 - y_1)^2}$

If you think about `x` and `y` beeing the attributes `Age` and the `Salary`, you will quickly see the underlying problem of this example. If we compute any two values from `Age` and the `Salary`, then `Salary` values will dominate the `Age` values, and it will produce an incorrect result. So to remove this issue, we need to perform feature scaling for machine learning.

There are two main ways to perform feature scaling in machine learning:
- Standardization and
- Normalization.

Standardization of a data point `x` can be calculated by using the mean $\bar x$ and the standard deviation $\sigma$ of the data set: <br>
$ x_{standardized} = \dfrac{x - \bar x}{\sigma}$

Normalization of a data point `x` can be calculated by using the minimum $min(x)$ and the maximum $max(x)$ of the data set: <br>
$ x_{normalized} = \dfrac{x - min(x)}{max(x) - min(x)}$

The Scikit-Learn `preprocessing` module provides the necessary methods for feature scaling. You can use the `StandardScaler()` method for Standardization and the `MinMaxScaler()` method for Normalization of attributes.

For example:

```python
# import StandardScaler method from the Preprocessing module of the Scikit-Learn library
from sklearn.preprocessing import StandardScaler

# make an instance of the StandardScaler class
scaler_standardization = StandardScaler()

# fit the data (calculate statistical features) and transform the data (apply statistical features for standardization)
x_train_standardized = scaler_standardization.fit_transform(x_train)

# displayed standardized training data set
x_train_standardized
```

```python
# import MinMaxScaler method from the Preprocessing module of the Scikit-Learn library
from sklearn.preprocessing import MinMaxScaler

# make a instance of the MinMaxScaler class
# feature_range -> scaling range, in this example we scale all data points between 0 and 1
scaler_normalization = MinMaxScaler(feature_range=(0,1))

# fit the data (calculate statistical features) and transform the data (apply statistical features for standardization)
x_train_normalized = scaler_normalization.fit_transform(x_train)

# displayed normalized training data set
x_train_normalized

# Change the variable feture_range for example to feature_range=(1,3) to see the difference
```
